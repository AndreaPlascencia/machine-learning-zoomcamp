Si el puntaje de información mutua es alto, significa que una variable proporciona información útil sobre la otra.En tu caso específico, mutual_info_score calcula la información mutua entre la variable redondeada rounded_snoring_rate y la variable categórica mrnaClassNum que indica si un paciente tiene un tumor o no. El puntaje resultante es un valor que cuantifica la relación entre estas dos variables en términos de información compartida. Un puntaje de información mutua de 0 indica que las variables son independientes, mientras que un puntaje más alto indica mayor dependencia o información compartida entre las variables.

## Feature Selection

```python
def bssWssFast(X, given_class_arr, num_class=2):
    class_vec = np.zeros((num_class, len(given_class_arr)))

    for k in range(1, num_class + 1):
        temp = np.zeros(len(given_class_arr))
        temp[given_class_arr.iloc[:, 0].values == (k - 1)] = 1
        class_vec[k - 1, :] = temp

    class_mean_arr = np.zeros(num_class)
    ratio = np.zeros(X.shape[1])

    for j in range(X.shape[1]):
        overall_mean = np.sum(X.iloc[:, j]) / len(X.iloc[:, j])

        for k in range(1, num_class + 1):
            class_mean_arr[k - 1] = np.sum(class_vec[k - 1, :] * X.iloc[:, j]) / np.sum(class_vec[k - 1, :])

        class_mean_vec = class_mean_arr[given_class_arr.iloc[:, 0].values]
        bss = np.sum((class_mean_vec - overall_mean) ** 2)
        wss = np.sum((X.iloc[:, j] - class_mean_vec) ** 2)

        # Manejo de división por cero
        ratio[j] = bss / wss if wss != 0 else np.nan

    selected_features = np.argsort(ratio)[::-1]
    return selected_features



bss = bssWssFast(mrnaData, mrnaClassNum, 2)
```

Selección de características de análisis discriminante basada en la relación entre la varianza entre clases y la varianza dentro de las clases. La función bssWssFast toma tres argumentos: X, given_class_arr, y num_class. X es una matriz de características, given_class_arr es una matriz de etiquetas de clase, y num_class es el número de clases. La función devuelve un vector de índices de características ordenados según su importancia para la clasificación.

El código comienza inicializando una matriz de ceros class_vec con dimensiones (num_class, len(given_class_arr)). Luego, se itera sobre cada clase y se crea un vector de ceros temp con longitud igual a la cantidad de elementos en given_class_arr. Los elementos de temp correspondientes a la clase actual se establecen en 1, y el vector resultante se almacena en la matriz class_vec. A continuación, se calcula la media de cada clase y se almacena en un vector class_mean_arr. Se calcula la media general de todas las características y se almacena en overall_mean. Luego, se itera sobre cada característica y se calcula la relación entre la varianza entre clases y la varianza dentro de las clases. Los índices de características se ordenan según su importancia para la clasificación y se devuelven como resultado.

```python
mrnaDataReduced = mrnaData.iloc[:, bss[:10]]
```

```python
# División en conjunto de prueba y conjunto temporal combinado
mrnaDataReduced_full_train, mrnaDataReduced_test, mrnaClassNum_full_train, mrnaClassNum_test = train_test_split(
    mrnaDataReduced, mrnaClassNum, test_size=0.2, random_state=1, stratify=mrnaClassNum
)

# División del conjunto temporal combinado en conjuntos de entrenamiento y validación
mrnaDataReduced_train, mrnaDataReduced_val, mrnaClassNum_train, mrnaClassNum_val = train_test_split(
    mrnaDataReduced_full_train, mrnaClassNum_full_train, test_size=0.25, random_state=1, stratify=mrnaClassNum_full_train
)

# Restablecer los índices
mrnaDataReduced_train = mrnaDataReduced_train.reset_index(drop=True)
mrnaDataReduced_val = mrnaDataReduced_val.reset_index(drop=True)
mrnaDataReduced_test = mrnaDataReduced_test.reset_index(drop=True)

# Las variables objetivo
y_train = mrnaClassNum_train.values.ravel()
y_val = mrnaClassNum_val.values.ravel()
y_test = mrnaClassNum_test.values.ravel()
```

We can then list the genes by their name as provided in the 'mrnaNorm' dataset, where the first column is the list of gene names. The names have two codes separated by character '|':

HUGO code Entrez code

For example, FHL1|2273 is gene with HUGO code FHL1 and Entrez code 2273.

```python
genes = mrnaNorm.iloc[bss[:10], 0]
genes
```

```python
result_table = []
columns = list(genes)
genes = list(bss[:10])

# Con las características reducidas
for i in range(10):
    model = LogisticRegression(solver='liblinear', random_state=1)
    model.fit(mrnaDataReduced_train.iloc[:, [i]], y_train)

    y_pred = model.predict(mrnaDataReduced_val.iloc[:, [i]])

    accuracy = accuracy_score(y_val, y_pred)

    result_table.append([columns[i], genes[i], accuracy])
```

```python
result_table = pd.DataFrame(result_table, columns=['Feature', 'Index', 'Accuracy'])
result_table
```

```python
features = [result_table.Index[0]]
limit = result_table.Accuracy[0]

for i in range(1, 10):
    features.append(result_table.iloc[i, 1])
    data = mrnaDataReduced_train.loc[:, features]
    model = LogisticRegression(solver='liblinear', random_state=1)
    model.fit(data, y_train)
    
    predict = model.predict(mrnaDataReduced_val.loc[:, features])
    
    accuracy = accuracy_score(y_val, predict)

    print(features, accuracy)
    
    if accuracy > limit:
        limit = accuracy
    else:
        del features[-1]

print(len(features))
print(limit)
```
