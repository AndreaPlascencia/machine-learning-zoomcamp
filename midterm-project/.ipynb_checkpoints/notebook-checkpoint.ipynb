```python
# Calcula el puntaje de información mutua
info_mutua_score = mutual_info_score(round(mrnaData.iloc[:, 4]), mrnaClassNum.values.ravel())

# Imprime el resultado
print(f"Puntaje de Información Mutua: {info_mutua_score}")
```

Si el puntaje de información mutua es alto, significa que una variable proporciona información útil sobre la otra.En tu caso específico, mutual_info_score calcula la información mutua entre la variable redondeada rounded_snoring_rate y la variable categórica mrnaClassNum que indica si un paciente tiene un tumor o no. El puntaje resultante es un valor que cuantifica la relación entre estas dos variables en términos de información compartida. Un puntaje de información mutua de 0 indica que las variables son independientes, mientras que un puntaje más alto indica mayor dependencia o información compartida entre las variables.

## Model Selection

```python
# mrnaData es el DataFrame y mrnaClassNum es la variable objetivo
# División en conjunto de prueba y conjunto temporal combinado
mrnaData_full_train, mrnaData_test, mrnaClassNum_full_train, mrnaClassNum_test = train_test_split(
    mrnaData, mrnaClassNum, test_size=0.2, random_state=1, stratify=mrnaClassNum
)

# División del conjunto temporal combinado en conjuntos de entrenamiento y validación
mrnaData_train, mrnaData_val, mrnaClassNum_train, mrnaClassNum_val = train_test_split(
    mrnaData_full_train, mrnaClassNum_full_train, test_size=0.25, random_state=1, stratify=mrnaClassNum_full_train
)

# Restablecer los índices
X_train = mrnaData_train.reset_index(drop=True)
X_val = mrnaData_val.reset_index(drop=True)
X_test = mrnaData_test.reset_index(drop=True)

# Las variables objetivo
y_train = mrnaClassNum_train.values.ravel()
y_val = mrnaClassNum_val.values.ravel()
y_test = mrnaClassNum_test.values.ravel()
```

```python
len(mrnaData_train), len(mrnaData_test), len(mrnaData_val), len(y_train), len(y_test), len(y_val)
```

## Model

```python
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(mrnaData_train, y_train)
```

```python
y_pred = model.predict_proba(mrnaData_val)[:, 1]
tumor_decision = (y_pred >= 0.5)
(y_val == tumor_decision).mean()
```

```python
y_pred = model.predict(mrnaData_val)
confusion_matrix(y_val, y_pred)
```

Explicar matriz

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
```

```python
X_train = mrnaData_train
```

```python
depths = [1, 2, 3, 10, 20, 30, 40, 50, 100, None]

for depth in depths: 
    dt = DecisionTreeClassifier(max_depth=depth)
    dt.fit(X_train, y_train)
    
    y_pred = dt.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_pred)
    
    print('%4s -> %.3f' % (depth, auc))
```

```python
scores = []

for depth in [2, 3, 10]:
    for s in [1, 5, 10, 15, 20, 500, 100, 200]:
        dt = DecisionTreeClassifier(max_depth=depth, min_samples_leaf=s)
        dt.fit(X_train, y_train)

        y_pred = dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)
        
        scores.append((depth, s, auc))
```

```python
columns = ['max_depth', 'min_samples_leaf', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
df_scores_pivot = df_scores.pivot(index='min_samples_leaf', columns=['max_depth'], values=['auc'])
df_scores_pivot.round(3)
```

```python
sns.heatmap(df_scores_pivot, annot=True, fmt=".3f")
```

```python
dt = DecisionTreeClassifier(max_depth=2, min_samples_leaf=20)
dt.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
scores = []

for n in range(10, 201, 10):
    rf = RandomForestClassifier(n_estimators=n, random_state=1)
    rf.fit(X_train, y_train)

    y_pred = rf.predict_proba(X_val)[:, 1]
    auc = roc_auc_score(y_val, y_pred)
    
    scores.append((n, auc))
```

```python
df_scores = pd.DataFrame(scores, columns=['n_estimators', 'auc'])
```

```python
plt.plot(df_scores.n_estimators, df_scores.auc)
```

```python
scores = []

for d in [2, 10, 15]:
    for n in range(10, 201, 10):
        rf = RandomForestClassifier(n_estimators=n,
                                    max_depth=d,
                                    random_state=1)
        rf.fit(X_train, y_train)

        y_pred = rf.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)

        scores.append((d, n, auc))
```

```python
columns = ['max_depth', 'n_estimators', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
for d in [2, 10, 15]:
    df_subset = df_scores[df_scores.max_depth == d]
    
    plt.plot(df_subset.n_estimators, df_subset.auc,
             label='max_depth=%d' % d)


plt.legend()
```

```python
max_depth = 10
```

```python
scores = []

for s in [1, 3, 5, 10, 50]:
    for n in range(10, 201, 10):
        rf = RandomForestClassifier(n_estimators=n,
                                    max_depth=max_depth,
                                    min_samples_leaf=s,
                                    random_state=1)
        rf.fit(X_train, y_train)

        y_pred = rf.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred)

        scores.append((s, n, auc))
```

```python
columns = ['min_samples_leaf', 'n_estimators', 'auc']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
colors = ['black', 'blue', 'orange', 'red', 'grey']
values = [1, 3, 5, 10, 50]

for s, col in zip(values, colors):
    df_subset = df_scores[df_scores.min_samples_leaf == s]
    
    plt.plot(df_subset.n_estimators, df_subset.auc,
             color=col,
             label='min_samples_leaf=%d' % s)

plt.legend()
```
